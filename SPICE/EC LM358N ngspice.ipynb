```python
import os
import time
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# oamp = 'LM2904N'
oamp = ['LM358N','LM393AT']
# Напряжение питания ОУ
oamp_pwr = ["3.3","5","12"]
# Импульс 20-200мкс
pulse = [p*20 for p in range(11) if p != 0]

# 1. Generate the Cartesian product (an iterator of tuples)
product_iterator = itertools.product(oamp, oamp_pwr, pulse)

# 2. Convert the iterator to a list and pass it to the DataFrame constructor
# You can optionally specify column names
vars = pd.DataFrame(list(product_iterator), columns=['oamp', 'oamp_pwr', 'pulse'])
vars['protect'] = ''
vars['protect_time'] = ''
vars['hi_volt'] = ''

# vars = vars.head(1)
# print(vars)
```

```python
%cd /content/res/

def run_ngspice(filename):
  !ngspice -b $filename".cir" 2>  $filename"_err.log" > /dev/null


def calc(id, var):
  # Каждая новая симуляция со своей временной меткой (именем фала)
  shem = str(time.time_ns())
  results = shem +'.res'

  control = f"""
  .control
  * Run the simulation
  tran 100n 0.5m
  * Write specific vectors
  wrdata {results} v(in) v(out)
  * Optional: exit ngspice automatically after script finishes in batch mode
  *quit
  .endc
  """

  open_collector = (True if 'LM393' in var['oamp']  else False)

  cir = f"""
  .title
  .include ../NATBJT.LIB
  .include ../TICOMP.LIB
  ************************************
  V1 in 0 DC 0V PULSE(0V 0.36V 0s 1e-07s 1e-07s {var['pulse']}us {var['pulse']*2}us)
  Vref_v ref_v 0 0.35V
  Vpwr pwr_in 0 {var['oamp_pwr']}
  Xop ref_v in pwr_in 0 out {var['oamp']}
  {'Roc out pwr_in 2k' if open_collector else ''}
  {'Rload out 0 1k' if not open_collector else ''}
  ************************************
  {control}
  ************************************
  .end
  """

  with open(shem+'.cir', "w") as file:
    file.write(cir)
  run_ngspice(shem)

  sim = pd.read_csv(results, sep=r'\s+')
  # удаляем нечетные колонки кроме первой (время)
  sim = sim.iloc[:, [0] + list(range(1, len(sim.columns), 2))]
  sim.columns = ['t', 'v(in)', 'v(out)']


  # Plotting the results.
  t = sim['t']*1e6
  plt.figure(figsize=(10, 4))
  plt.plot(t, sim['v(in)'], color='blue', label='v(in)', linewidth=0.8)
  plt.plot(t, sim['v(out)'], color='red', label='v(out)', linewidth=0.8)
  plt.title(f'{var['oamp']} импульс на инверсном входе {var['pulse']}мкс при Uпит={var['oamp_pwr']}')
  plt.xlabel('Время, мкс')
  plt.ylabel('Напряжение, в')
  plt.xlim(-1, 120)
  # plt.ylim(-1, 2)
  plt.grid(True)
  plt.tight_layout()
  plt.legend()
  plt.show()


  # Получаем 'стартовое' напряжение
  vars.loc[id,'hi_volt'] = sim ['v(out)'].iloc[0]

  # print(var)
  # Отлавливаем время когда напряжение упадет ниже 0.65В
  filtered_query = sim.query("`v(out)` < 0.65")
  if filtered_query.size == 0:
    protect_time = ''
    protect = '' # "Защита не сработала. Напряжение не опустилось ниже порога."
  else:
    protect_time = round(filtered_query['t'].iloc[0]*1e6,2)
    protect = '✓' #f'время реакции={protect_time} мкс'
  vars.loc[id,'protect'] = protect
  vars.loc[id,'protect_time'] = protect_time
  display(var)

!rm -rf /content/res/*
for id, var in vars.iterrows():
  calc(id, var)
```
